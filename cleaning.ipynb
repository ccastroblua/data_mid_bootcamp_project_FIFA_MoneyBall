#### Cleaning function for 'nationality' variable:

```python
def cleaning_nationality(country):
    top25 = ['England', 'Germany', 'Spain', 'France', 'Brazil', 'Argentina', 'Italy', 'Netherlands', 'Portugal', 'United States',
            'Uruguay', 'Mexico', 'Colombia', 'Japan', 'Poland', 'Scotland', 'Republic of Ireland', 'Belgium', 'Sweden', 'Norway',
            'Denmark', 'Austria', 'Turkey', 'Switzerland', 'Paraguay']
    
    if country in top25:
        country = country.lower().replace(' ','_')
    else:
        country = 'other'
    
    return country
```

#### Cleaning function for 'height' variable:

```python
def cleaning_height(height):
    converted_height = float(height.split("'")[0])*30.48 + float(height.split("'")[1][:-1])*2.54
    return converted_height
```

#### Cleaning function for 'weight' variable:

```python
def cleaning_weight(weight):
    converted_weight = float(weight[:-3])*0.4536
    return converted_weight
```

#### Cleaning function for 'wage' and 'release_clause':

```python
def cleaning_money(money):
    if money[-1:] == '0':
        return 0
    elif money[-1:] == 'K':
        return float(money[1:-1])*1000
    elif money[-1:] == 'M':
        return float(money[1:-1])*1000000
```

#### Cleaning function for 'w/f', 'sm', 'ir':

```python
def cleaning_stars(stars):
    return int(stars[:1])-1
```

#### Cleaning function for 'a/w', 'd/w':

```python
def cleaning_rating(rating):
    if rating == 'High':
        return 2
    elif rating == 'Medium':
        return 1
    else:
        return 0
```

#### Cleaning function for 'hits':

```python
def cleaning_hits(hits):
    if type(hits) == int:
        return float(hits)
    else:
        if hits[-1:] == 'K':
            return float(hits[:-1])*1000
        else:
            return float(hits)
```

#### Cleaning function for all the skills:

```python
def cleaning_skills(skill):
    return int(skill.split('+')[0])
```

#### Master cleaning function:

```python
def cleaning_categoricals(df):
    df['nationality'] = df['nationality'].apply(cleaning_nationality)
    df.drop('club', axis=1, inplace=True)
    df['height'] = df['height'].apply(cleaning_height)
    df['weight'] = df['weight'].apply(cleaning_weight)
    df['release_clause'] = df['release_clause'].apply(cleaning_money)
    df['wage'] = df['wage'].apply(cleaning_money)
    df['value'] = df['value'].apply(cleaning_money)
    df['w/f'] = df['w/f'].apply(cleaning_stars)
    df['sm'] = df['sm'].apply(cleaning_stars)
    df['ir'] = df['ir'].apply(cleaning_stars)
    df['a/w'] = df['a/w'].apply(cleaning_rating)
    df['d/w'] = df['d/w'].apply(cleaning_rating)
    df['hits'] = df['hits'].apply(cleaning_hits)
    
    skills_columns = ['ls', 'st', 'rs', 'lw', 'lf', 'cf', 'rf', 'rw', 'lam', 'cam', 'ram', 'lm', 'lcm',
    'cm', 'rcm', 'rm', 'lwb', 'ldm', 'cdm', 'rdm', 'rwb', 'lb', 'lcb', 'cb',
    'rcb', 'rb', 'gk']
    
    for col in skills_columns:
        df[col] = df[col].apply(cleaning_skills)
    
    return df
```
