```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import Normalizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
import pickle
```

```python
pd.set_option('max_columns', None)
pd.set_option('max_rows', 100)
```

```python
data = pd.read_csv('./data/cleaned_data.csv', index_col='id')
```

```python
data.head()
```

```python
y = data['ova']
X = data.drop('ova', axis=1)
```

```python
print(y.shape, X.shape)
```

```python
X_cat = X.select_dtypes(include=object)
X_num = X.select_dtypes(exclude=object)
```

```python
print(X_cat.shape, X_num.shape)
```

#### Let's start transforming categoricals variables:

```python
X_cat.head()
```

```python
print(X_cat['nationality'].nunique())
print(X_cat['bp'].nunique())
```

```python
# Both variables do not have an specific order of importance that we can apply with numbers, so I will use One Hot Encoder:

encoder = OneHotEncoder(handle_unknown='error', drop='first')
encoder.fit(X_cat)
```

```python
encoded = encoder.transform(X_cat).toarray()
print(encoded)
```

```python
encoded.shape
```

```python
categories_names = list(encoder.categories_[0][1:])+list(encoder.categories_[1][1:])
```

```python
X_cat_encoded = pd.DataFrame(encoded)
X_cat_encoded.columns = categories_names
X_cat_encoded
```

#### Correlation Matrix check:

```python
X_num
```

```python
correlation_data = data.select_dtypes(exclude=object)
```

```python
corr_matrix = correlation_data.corr()
```

```python
corr_matrix[corr_matrix['ova'] >= 0.6]
```

```python
corr_matrix[corr_matrix['ova'] <= -0.6]
```

#### Tranforming numericals variables

```python
# Only keeping 'bov', 'reactions', 'composure', 'total_stats', 'base_stats', 'pas', 'dri', 'value', 'release_clause' and 'growth':
X_num_columns = ['bov', 'composure', 'pas', 'value', 'growth']
X_num_v2 = X_num[X_num_columns]
```

```python
sns.heatmap(X_num_v2.corr(), annot=True)
```

```python
# Have to remove 'release_clause' due to the correlation with 'value'. 
# Also 'reaction', 'base_stats', 'total_stats', 'dri'.
```

```python
X_num_v2
```

#### Now let's transform this numerical variables:

```python
scaler = MinMaxScaler()
scaler.fit(X_num_v2)
print(scaler)
```

```python
# Minimum and Maximum with the MixMaxScaler:

print(scaler.data_max_, scaler.data_min_)
```

```python
X_num_minmax = scaler.transform(X_num_v2)
```

```python
X_num_minmax = pd.DataFrame(X_num_minmax)
X_num_minmax.columns = ['bov','composure','pas','value','growth']
X_num_minmax
```

#### Other method Normalizer:

```python
transformer = Normalizer()
transformer.fit(X_num_v2)
x_normalized = transformer.transform(X_num_v2)
x_normalized.shape
```

```python
X_num_normalized = pd.DataFrame(x_normalized)
```

```python
X_num_normalized.columns = ['bov','composure','pas','value','growth']
X_num_normalized
```

#### Reseting index from y:

```python
y = y.reset_index()
```

```python
y.drop('id', axis=1 ,inplace=True)
```

```python
y
```

#### Let's concat categoricals and numericals (using MinMaxScaler):

```python
X_transformed = pd.concat([X_cat_encoded, X_num_minmax], axis=1)
```

```python
X_transformed.shape
```

```python
X_transformed.head()
```

#### Train, test split:

```python
X_train, X_test, y_train, y_test = train_test_split(X_transformed, y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

#### Training the model:

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

#### Analizing results:

```python
predictions = lr.predict(X_test)
```

```python
r2_score(y_test, predictions), mean_absolute_error(y_test, predictions), mean_squared_error(y_test, predictions, squared=False)
```
