#### Standarizing columns names:

```python
def rename_columns(dataframe):
    cols = dataframe.columns
    final_columns = []
    
    for c in cols:
        final_columns.append(c.lower().replace(' ','_'))
        
    dataframe.columns = final_columns
    return dataframe
```

```python
data1 = rename_columns(data.copy())
```

```python
data1.columns
```

#### Set indexes player_id:

```python
data1.set_index('id', inplace=True)
```

#### Let's drop the next columns:

#### Let's try to understand which columns won't need in our first iteration:
- __id__: We will use it as our index.
- __name__: Your birthname won't help you to be a better player).
- __player_photo__: Not useful
- __club_logo__: Not useful
- __flag_photo__: Not useful
- __team_&_contract__: Contract duration is not quite usefull. We will discard this column in the first iteration.
- __joined__: Not necesary for now.
- __loan_date_end__: Not necesary for now and full of nulls.
- __contract__: Not necesary for now.
- __gender__: Is useless. All values are 'Male'.
- __position__: We will drop this one and use only 'bp' (best_position)

```python
drop_columns = ['name', 'player_photo', 'club_logo', 'flag_photo', 'team_&_contract', 'joined', 'loan_date_end', 'contract', 'gender']
```

```python
data1.drop(labels=drop_columns, axis=1, inplace=True)
```

#### Checking nulls:

```python
data1.isna().sum()
```

```python
data1.shape
```

#### Let's get a little deeper on the nulls that we stills have:

```python
data1[data1['composure'].isna() & data1['position'].isna()]
```

#### We have 140 rows that share nulls between 'composure' and 'position'. If we drop both columns null rows we will drop only 696 (only 4% of the rows).

```python
data1.dropna(subset=['composure', 'position'], inplace=True)
```

```python
data1.isna().sum()
```

#### Second column selection:
- __foot__: Let's drop it for our first iteration. It shouldn't affect our final result.
- __position__: We will drop this one and use only 'bp' (best_position)

```python
drop_columns_v2 = ['foot', 'position']
data1.drop(labels=drop_columns_v2, axis=1, inplace=True)
```

```python
data1.shape
```

#### With that cleaning we drop almost all the nulls in all columns. We only have 15 nulls in 'club' column. Let's check them:

```python
data1[data1['club'].isna()]
```

#### All of these players are free agents. Let's remove them from our dataset:

```python
data1.dropna(subset=['club'], inplace=True)
```
